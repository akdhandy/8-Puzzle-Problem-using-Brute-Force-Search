```python
import numpy as np
import cv2

#check if the coordinate is in the cirlce obstacle
def circle(x,y):
    #checks if the coordinate is within the geometrical equation
    if ((x-225)**2+(y-50)**2) < (25)**2:
        return True
    else:
        return False

#check if the coordinate is in the ellipse obstacle
def ellipse(x,y):
    #checks if the coordinate is within the geometrical equation
    if ((x-150)**2.0)/(40)**2 + ((y-100)**2.0)/(20)**2 <= 1:
        return True
    else:
        return False

def polygon(x,y):
    #points of the polygon
    p1 = [25.0, 15]
    p2 = [75.0, 15]
    p3 = [100.0, 50]
    p4 = [75.0, 80]
    p5 = [50.0, 50]
    p6 = [20.0, 80]


    #two separate shapes
    #first shape
    m6 = (p1[1]-p6[1])/(p1[0]-p6[0])
    b6 = p6[1]-m6*p6[0]

    m1 = (p2[1]-p1[1])/(p2[0]-p1[0])
    b1 = p1[1]-m1*p1[0]

    m2 = (p2[1]-p5[1])/(p2[0]-p5[0])
    b2 = p2[1]-m2*p2[0]

    m56 = (p5[1]-p6[1])/(p5[0]-p6[0])
    b56 = p2[1]-m56*p2[0]

    #second shape
    m22 = (p2[1]-p3[1])/(p2[0]-p3[0])
    b22 = p2[1]-m22*p2[0]

    m3 = (p3[1]-p4[1])/(p3[0]-p4[0])
    b3 = p3[1]-m3*p3[0]

    m4 = (p4[1]-p5[1])/(p4[0]-p5[0])
    b4 = p4[1]-m4*p4[0]

    m5 = (p2[1]-p5[1])/(p2[0]-p5[0])
    b5 = p5[1]-m5*p5[0]

    #make inequalities with the lines so that bounds make up the shape of the obstacle
    #combines thet two shapes above
    if  (m1*x + b1 < y and m2*x + b2 > y and m6*x + b6 < y and m56*x + b56) or (m22*x + b22 < y and m3*x + b3 > y and m4*x + b4 > y and m5*x + b5 < y):
        return True
    else:
        return False

    
def rhombus(x,y):
    #create the points of the diamond
    p11 = [225.0, 160.0]
    p22 = [250.0, 175.0]
    p33 = [225.0, 190.0]
    p44 = [200.0, 175.0]

    #Making lines connecting the points with the help of line equation
    #m -slope 
    #b is the y-intercept
    m11 = (p22[1]-p11[1])/(p22[0]-p11[0])
    b11 = p11[1]-m11*p11[0]

    m22 = (p22[1]-p33[1])/(p22[0]-p33[0])
    b22 = p22[1]-m22*p22[0]

    m33 = (p33[1]-p44[1])/(p33[0]-p44[0])
    b33 = p33[1]-m33*p33[0]

    m44 = (p11[1]-p44[1])/(p11[0]-p44[0])
    b44 = p44[1]-m44*p44[0]
    
    #Making up the shape of the obstacle with inequalities with respect to half planese
    if m11*x + b11 < y and m22*x + b22 > y and m33*x + b33 > y and m44*x + b44 < y:
        return True
    else:
        return False 
    
def rectangle(x,y):
    #Rectangle coordinates
    p1 = [39.0, 127]
    p2 = [104.0, 165]
    p3 = [95.0, 170]
    p4 = [30.0, 132]

    #Making lines connecting the points with the help of line equation
    #m -slope 
    #b is the y-intercept
    m1 = (p2[1]-p1[1])/(p2[0]-p1[0])
    b1 = p1[1]-m1*p1[0]

    m2 = (p2[1]-p3[1])/(p2[0]-p3[0])
    b2 = p2[1]-m2*p2[0]

    m3 = (p3[1]-p4[1])/(p3[0]-p4[0])
    b3 = p3[1]-m3*p3[0]

    m4 = (p1[1]-p4[1])/(p1[0]-p4[0])
    b4 = p4[1]-m4*p4[0]

    #Making up the shape of the obstacle with inequalities with respect to half planes
    if m1*x + b1 < y and m2*x + b2 > y and m3*x + b3 > y and m4*x + b4 < y:
        return True
    else:
        return False

    
# Maze Initialization
height = 200
width = 300

mask = (np.full([height,width], 0, dtype='uint8'))

#Display obstacles
for i in range(200):
    for j in range(300):
        if circle(j,i) or ellipse(j,i) or polygon(j,i) or rectangle(j,i) or rhombus(j,i):
            #if the coordinates are within an obstacle make that pixel lighter
            mask[i][j] = 255

cv2.imshow("Map",mask)
cv2.waitKey(0)
# cleanup the camera and close any open windows
cv2.destroyAllWindows()
```
