```python

import numpy as np
import os
import copy

initial_state = []


def get_initial():
    print("Enter number from 0-8")
    initial_state = np.zeros(9)
    for i in range(9):
        states = int(input("Enter the number: "))
        initial_state[i] = np.array(states)
    print(initial_state)
    initial_state = np.reshape(initial_state,(3,3))
    return initial_state


def find_index(puzzle):
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] == 0:
                return i,j

#def blank_tile_location(A):
    #loc = np.argwhere(A == 0)
    #return loc            
            

def move_left(data):
    temp_arr = copy.deepcopy(data)
    i, j = find_index(temp_arr)
    if j != 0:
        temp = temp_arr[i, j - 1]
        temp_arr[i, j] = temp
        temp_arr[i, j - 1] = 0
        print("Left motion")
        return temp_arr
    else:
        print("No left motion")
        return temp_arr


def move_right(data):
    temp_arr = copy.deepcopy(data)
    i, j = find_index(data)
    if j != 2:
        temp = temp_arr[i, j + 1]
        temp_arr[i, j] = temp
        temp_arr[i, j + 1] = 0
        print("Right motion")
        return temp_arr
    else:
        print("No right motion")
        return temp_arr

def move_up(data):
    temp_arr = copy.deepcopy(data)
    i, j = find_index(data)
    if i != 0:
        temp_arr = np.copy(data)
        temp = temp_arr[i - 1, j]
        temp_arr[i, j] = temp
        temp_arr[i - 1, j] = 0
        print("Up motion")
        return temp_arr
    else:
        print("No up motion")
        return temp_arr


def move_down(data):
    temp_arr = copy.deepcopy(data)
    i, j = find_index(data)
    if i != 2:
        temp_arr = np.copy(data)
        temp = temp_arr[i + 1, j]
        temp_arr[i, j] = temp
        temp_arr[i + 1, j] = 0
        print("Down motion")
        return temp_arr
        
    else:
        print("No down motion")
        return temp_arr

#to check and append in the parent node
def check_and_append(p_node, new_node, count):
    stat=False
    j=1
    for i in range(count):
        m=0
        for l in range(3):
            for k in range(3):
                if (p_node[i][l][k]==new_node[l][k]):  
                    m+=1
        if m==9:
            j=0
    print(j)    
    if j==1:
        p_node.append(new_node)
        count+=1
        stat=True
    return p_node, count, stat

# to check if the goal is reached
def goal_check(B,goal_node):
    status = np.array_equal(B,goal_node)
    return status

# initializing the goal node
goal_node = np.array([[1,2,3],[8,0,4],[7,6,5]])

child_index=[]
initial_array = get_initial()
child_index.append(initial_array)
child_node_number=1
print("child index total")
print(child_index)
print("child index 1.")
print(child_index[0])
temp_index = []
i=0
while (i<child_node_number):

    node_list=child_index[i]
    
    new_node = move_up(node_list)
    print(new_node)
    child_index, child_node_number, status=check_and_append(child_index, new_node, child_node_number)
    print(status)
    if status == True:
        child_node_number += 1
        temp_index.append(child_node_number)
        child_index.append(new_node)
       # temp_node_info = np.array([child_node_number,i,0])  #
       # node_info.append(temp_node_info)#
            

    if goal_check(new_node,goal_node):
        found = True
        #goal_index = child_node_number
        break

    new_node = move_down(node_list)
    print(new_node)
    child_index, child_node_number, status=check_and_append(child_index, new_node, child_node_number)
    print(status)
    if status == True:
        child_node_number += 1
        child_index.append(new_node)
        #temp_node_info = np.array([child_node_number,i,0])#
        #node_info.append(temp_node_info)#
        temp_index.append(child_node_number)
            
    
    if goal_check(new_node,goal_node):
        found = True
        #goal_index = child_node_number
        break

    new_node = move_left(node_list)
    print(new_node)
    child_index, child_node_number, status=check_and_append(child_index, new_node, child_node_number)
    print(status)
    if status == True: 
        child_node_number += 1
        child_index.append(new_node)
        #temp_node_info = np.array([child_node_number,i,0])
        #node_info.append(temp_node_info)
        temp_index.append(child_node_number)
            
    
    if goal_check(new_node,goal_node):
        found = True
        #goal_index = child_node_number
        break
    
    new_node = move_right(node_list)
    print(new_node)
    child_index, child_node_number, status=check_and_append(child_index, new_node, child_node_number)
    print(status)
    if status == True:
        child_node_number += 1
        child_index.append(new_node)
        #temp_node_info = np.array([child_node_number,i,0])           
        #node_info.append(temp_node_info)
        temp_index.append(child_node_number)
            
    if goal_check(new_node,goal_node):
        found = True
        break
    print(child_index)    
    print(child_node_number)
    i+=1


```
